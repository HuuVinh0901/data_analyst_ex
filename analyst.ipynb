```python
# Câu 3
import pandas as pd
column_name=['Id','Name','Age','Weight','m0006','m0612','m1218','f0006','f0612','f1218']
df= pd.read_csv('patient_heart_rate.csv')
df.head(10)
```

```python
# Câu 4
df[['Firstname', 'Lastname']] = df['Name'].str.split(expand=True) 
df = df.drop('Name', axis=1)
```

```python
# Câu 5
weight = df['Weight']
for i in range(0, len (weight)): 
    x = str(weight[i]) 
    if 'lbs' in x[-3:]:
        x = x[:-3:]
        float_x = float(x)
        y = int(float_x/2.2) #y = str(y) weight[i]=y
    if 'kgs' in x[-3:]: 
        x = x[:-3:]
        float_x = float(x)
    weight[i]=x
df.rename(columns={'Weight':'Weight_kgs'},inplace=True)
df
```

```python
# Câu 6
df.dropna(how='all',inplace=True)
```

```python
# Câu 7
df =df.drop_duplicates(subset=['Firstname', 'Lastname', 'Age', 'Weight_kgs'])
```

```python
# Câu 8

df.Firstname.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True) 
df.Lastname.replace({r'[^\x00-\x7F]+':''}, regex=True, inplace=True)
print (df)
```

```python
# Câu 9
df['Age'].fillna(df['Age'].mode()[0],inplace=True)
```

```python
# Câu 10
df['Weight_kgs'].fillna(df['Weight_kgs'].median,inplace=True)
```

```python
# Câu 11
# B1
df = pd.melt (df, id_vars=['Id', 'Age', 'Weight_kgs', 'Firstname', 'Lastname'], value_name='PulseRate', var_name='gender_time').sort_values(['Id', 'Age', 'Weight_kgs', 'Firstname', 'Lastname'])
```

```python
# B2
df_temp=df['gender_time'].str.extract("(\D)(\d+)(\d{2})",expand=True)
df_temp.head()
```

```python
# B3
df_temp.columns=['Gender','Lower_hour','Upper_hour']
df_temp.head()
```

```python
# B4
df=pd.concat([df,df_temp],axis=1)
```

```python
# B5
df=df.drop(['gender_time'],axis=1)
```

```python
# Câu 12
import numpy as np
df=df.replace('-',np.nan).dropna(subset=['PulseRate'])
```

```python
# Câu 14
df=df.reset_index()
```

PHÂN TÍCH DỮ LIỆU

```python
df = pd.read_csv('patient_heart_rate_clean.csv')
```

```python
import matplotlib.pyplot as plt
```

```python

# Câu 1
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(
    gender_counts,
    labels=gender_counts.index,
    autopct='%1.1f%%',
    startangle=140,
    shadow=True
)
plt.title('Tỉ lệ bệnh nhân nam và nữ')
plt.show()
```

```python
# Câu 2
df['Fullname'] = df['Firstname'] + ' ' + df['Lastname']
df
```

```python
import seaborn as sns
```

```python
# Câu 3

plt.figure(figsize=(10, 6))
sns.barplot(data=df,x='Fullname', y='Age')
plt.title('So sánh độ tuổi giữa các tên bệnh nhân')
plt.xticks(rotation=90)
plt.show()
```

```python
# 4. Vẽ biểu đồ cột so sánh nhịp tim trung bình giữa các tên bệnh nhân 
PulseRate_comparison = df.groupby('Fullname')['PulseRate'].mean().reset_index()
PulseRate_comparison = PulseRate_comparison.sort_values(by='PulseRate')
plt.figure(figsize=(10, 6))
sns.barplot(x='PulseRate', y='Fullname', data=PulseRate_comparison, palette='viridis')
plt.title('So sánh nhịp tim trung bình giữa các tên bệnh nhân')
plt.xlabel('Nhịp tim')
plt.ylabel('Tên Bệnh nhân')
plt.show()
```

```python
# 5. Vẽ biểu đồ cột so sánh nhịp tim trung bình theo từng nhóm giờ 00-06, 06-12,12-18 có phân loại theo giới tính
df['TimeGroup'] = df['Lower_hour'].astype(str) + '-' + df['Upper_hour'].astype(str)
heart_rate_grouped = df.groupby(['TimeGroup', 'Gender'])['PulseRate'].mean().reset_index()

sns.barplot(x='TimeGroup', y='PulseRate', hue='Gender', data=heart_rate_grouped)
plt.title('So sánh nhịp tim trung bình theo từng nhóm giờ và giới tính')
plt.show()
```

```python
age_heart_rate_corr = df.groupby('Fullname').agg({'Age': 'mean', 'PulseRate': 'mean'}).reset_index()
plt.scatter(age_heart_rate_corr['Age'], age_heart_rate_corr['PulseRate'])
plt.title('Tương quan giữa độ tuổi trung bình và nhịp tim trung bình')
plt.xlabel('Độ tuổi trung bình')
plt.ylabel('Nhịp tim trung bình')
plt.show()
```

```python
# 7. Vẽ biểu đồ đường thể hiện sự thay đổi của nhịp tim trung bình các bệnh nhân theo thứ tự 
# khung giờ 00-06, 06-12,12-18 theo từng nhóm giới tính.
sns.lineplot(x='TimeGroup', y='PulseRate', hue='Gender', data=heart_rate_grouped)
plt.title('Sự thay đổi nhịp tim trung bình theo giờ và giới tính')
plt.show()
```
